```python
#importing the library 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#reading the dataset from wikipedia website using pandas
%matplotlib notebook
pg = "https://en.wikipedia.org/wiki/Indian_cricket_team_records"
dfs_dd = pd.read_html(pg)
# dfs_dd[8].to_csv('ex.csv')
# df_win_loss = pd.read_csv('ex.csv')
# df_win_loss

dfs_dd[2][:-1]
```

```python
#reading the columns which are required for analysis
df_win_loss_dd = dfs_dd[2][:-1].copy()
cols = ['Opponent', '% Won']
win_loss_dd = df_win_loss_dd[cols]
win_loss_dd=win_loss_dd[1:][:-1]
```

```python
win_loss_dd=win_loss_dd.drop([12],axis=0)
```

```python
edition_list = []
win_per_list = []
edition_list = win_loss_dd['Opponent']
win_per_list = win_loss_dd['% Won']
```

```python
#object to float list
win= np.array(win_per_list).tolist()
win
win_percent =[float(i) for i in win]
```

```python
y_pos = np.arange(len(edition_list))
y_pos
```

```python
plt.bar(y_pos,win_percent, label="Delhi Capitals Win %")
plt.xticks(y_pos, edition_list,rotation=90)
plt.ylabel('Win %')

plt.xlabel('Teams')
plt.title('India Win % over different opponent')
plt.show()
```
